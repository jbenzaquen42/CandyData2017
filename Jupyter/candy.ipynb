```python
# Cleaning up column names
column_names = []

for col in data.columns:
    # Split the name by spaces to remove any duplicate terms
    terms = col.split()
    unique_terms = []
    for term in terms:
        if term not in unique_terms:
            unique_terms.append(term)
    corrected_name = ' '.join(unique_terms)
    
    # Ensure consistent naming for demographic information
    corrected_name = corrected_name.replace('Going Out? GOING OUT?', 'Going Out?')\
                                   .replace('Gender GENDER', 'Gender')\
                                   .replace('Age AGE', 'Age')\
                                   .replace('Country COUNTRY', 'Country')\
                                   .replace('State-Province STATE- PROVINCE- COUNTY- ETC', 'State/Province')
    
    column_names.append(corrected_name)

# Update the DataFrame with corrected column names
data.columns = column_names

# Display the first few rows of the DataFrame with updated column names for verification
data.head()
```
